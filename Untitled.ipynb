```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load the CSV file
file_path = 'ACN.csv'
data = pd.read_csv(file_path)

# Convert 'Date' column to datetime format
data['Date'] = pd.to_datetime(data['Date'])

# Ensure there are no missing values in the relevant columns
data = data.dropna(subset=['Close(ACN)', 'Close(S&P500)'])

# Calculate daily returns if not already present
if '_Returns_ACN' not in data.columns or 'Returns_S&p' not in data.columns:
    data['_Returns_ACN'] = data['Close(ACN)'].pct_change()
    data['Returns_S&p'] = data['Close(S&P500)'].pct_change()

# Drop rows with NaN values resulting from pct_change calculation
data = data.dropna(subset=['_Returns_ACN', 'Returns_S&p'])

# Check if data is sufficient for calculation
if len(data) < 2:
    raise ValueError("Insufficient data to calculate beta. Ensure your dataset has more than one row.")

# Calculate Beta Index
def calculate_beta(data):
    try:
        covariance_matrix = np.cov(data['_Returns_ACN'], data['Returns_S&p'])
        beta = covariance_matrix[0, 1] / covariance_matrix[1, 1]
        return beta
    except ZeroDivisionError:
        print("Error: Variance of S&P500 returns is zero.")
        return np.nan
    except Exception as e:
        print(f"Error calculating beta: {e}")
        return np.nan

beta = calculate_beta(data)
if not np.isnan(beta):
    print(f'Beta Index: {beta:.2f}')
else:
    print('Unable to calculate Beta Index due to errors.')

# Plot the ACN stock price
plt.figure(figsize=(12, 6))
plt.plot(data['Date'], data['Close(ACN)'], label='ACN Stock Price', color='blue')
plt.xlabel('Date')
plt.ylabel('Price')
plt.title('ACN Stock Price Over Time')
plt.legend()
plt.grid(True)

# Display Beta Index in the plot if valid
if not np.isnan(beta):
    plt.figtext(0.15, 0.85, f'Beta Index: {beta:.2f}', fontsize=12, bbox=dict(facecolor='white', alpha=0.5))

# Show the plot
plt.show()








```
